```python
# Since one hot ecoding needs a 2D, therefore,
# we will convert the pandas series into a 2D array
categorical_data_array=df['ocean_proximity'].values
```

```python
# reshape method is used to reshape an array
array_2d = categorical_data_array.reshape(-1,1)
```

```python
array_2d
```

```python
# Use of one hot encoder
# one hot encoding - It transforms each unique category of a column into a new binary column (0 or 1).
from sklearn.preprocessing import OneHotEncoder
```

```python
# creating object of a class
encoder = OneHotEncoder()
```

```python
# Fit_transform:- The fit methods calculates the requried parameters
# and the transform() methods applies the calculated parameters to the data
encoded_data= encoder.fit_transform(array_2d)
```

```python
# This returns a sparse matrix where majority of the values are zeros
encoded_data
```

```python
# To visualize the data, we convert it into an array
encoded_data.toarray()
```

```python
# This method is used to return the order of unique categories (in which the matrix is created)
encoder.categories_[0]
```

```python
# creating a new dataframe of binary columns and merging it with the original dataset
df[encoder.categories_[0]] = encoded_data.toarray()
```

```python
df.head()
```

```python
# As the data/information of the ocean_proximity column is preserved in the form of new binary
# columns, hence we can remove this from the dataset
df.drop('ocean_proximity',axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```
